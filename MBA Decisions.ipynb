```python
# 1. Importing the required libraries for EDA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)  
# 2. Loading the data into the data frame.
df = pd.read_csv(r"D:\Download\archive\mba_decision_dataset.csv")
df.head(5)
df.tail(5)
# 3. Checking the types of data
df.dtypes
df.shape
# 4. Dropping the duplicate rows
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
df.count()
# 5. Dropping the missing or null values.
print(df.isnull().sum())
df.drop_duplicates(inplace=True)
# 6. Detecting Outliers
sns.boxplot(x=df['Annual Salary (Before MBA)'])
sns.boxplot(x=df['Expected Post-MBA Salary'])
# 7. Analyzing factors effect to decisions
# Year of experience
plt.figure(figsize=(8, 5))
sns.boxplot(x="Decided to Pursue MBA?", y="Years of Work Experience", data=df)
plt.title("MBA Decision vs Work Experience")
plt.show()
# Salary before MBA
plt.figure(figsize=(8, 5))
sns.boxplot(x="Decided to Pursue MBA?", y="Annual Salary (Before MBA)", data=df)
plt.title("MBA Decision vs Pre-MBA Salary")
plt.show()
# GRE/GMAT Score
plt.figure(figsize=(8, 5))
sns.histplot(df[df["Decided to Pursue MBA?"] == "Yes"]["GRE/GMAT Score"], bins=20, color="blue", label="Yes", kde=True)
sns.histplot(df[df["Decided to Pursue MBA?"] == "No"]["GRE/GMAT Score"], bins=20, color="red", label="No", kde=True)
plt.title("GRE/GMAT Score Distribution for MBA Decision")
plt.legend()
plt.show()
# Major
plt.figure(figsize=(10, 5))
sns.countplot(x="Undergraduate Major", hue="Decided to Pursue MBA?", data=df)
plt.xticks(rotation=45)
plt.title("MBA Decision by Undergraduate Major")
plt.show()
# MBA Funding Source 
plt.figure(figsize=(8, 5))
sns.countplot(x="MBA Funding Source", hue="Decided to Pursue MBA?", data=df)
plt.title("MBA Decision by Funding Source")
plt.show()
```
